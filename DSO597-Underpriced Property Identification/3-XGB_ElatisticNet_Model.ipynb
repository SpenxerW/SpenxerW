```python
import pandas as pd
```

```python
from sklearn.metrics import * 
import numpy as np
```

```python
data = pd.read_csv('0305_property_data_imputed.csv', index_col = 0)
```

```python
data.info()
```

```python
data = data.dropna().copy()
```

```python
X = data.loc[:, data.columns != 'RENT_SQFT'].copy()
y = data['RENT_SQFT'].copy()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size=0.30, random_state=1)
```

```python
train_X = train.loc[:, data.columns != 'RENT_SQFT'].copy()
test_X = test.loc[:, data.columns != 'RENT_SQFT'].copy()
```

```python
train_y = train['RENT_SQFT'].copy()
test_y = test['RENT_SQFT'].copy()
```

```python
from xgboost.sklearn import XGBRegressor
from sklearn.model_selection import cross_val_score, GridSearchCV

xgb = XGBRegressor()
parameters = {'nthread':[4], 
              'objective':['reg:squarederror'],
              'learning_rate': [0.01, 0.05, 0.10], 
              'max_depth': [3, 5, 10],
              'min_child_weight': [5, 8, 10],
              'subsample': [0.5, 0.9],
              'colsample_bytree': [0.7],
              'n_estimators': [150, 200, 400]}

gsc_xgb = GridSearchCV(xgb,
                        parameters,
                        cv = 3,
                        n_jobs = 3,
                        verbose=True)

grid_result_xgb = gsc_xgb.fit(train_X, train_y.values.ravel())
best_params_xgb = grid_result_xgb.best_params_
```

```python
best_params_xgb
```

```python
xgb = XGBRegressor(nthread=best_params_xgb['nthread'], objective=best_params_xgb['objective'],
                   learning_rate= best_params_xgb['learning_rate'],
                   max_depth=best_params_xgb['max_depth'],
                   min_child_weight = best_params_xgb['min_child_weight'], 
                   subsample = best_params_xgb['subsample'],
                   colsample_bytree = best_params_xgb['colsample_bytree'],
                   n_estimators = best_params_xgb['n_estimators'])
xgb.fit(train_X, train_y.values.ravel())
```

```python
xgb_pre_train = xgb.predict(train_X)
rmse_xgb = np.sqrt(mean_squared_error(train_y, xgb_pre_train))
```

```python
rmse_xgb
```

```python
xgb_pre_test = xgb.predict(test_X)
rmse_xgb_test = np.sqrt(mean_squared_error(test_y, xgb_pre_test))
rmse_xgb_test
```

```python
X = data.loc[:, data.columns != 'RENT_SQFT'].copy()
pred_xgb = xgb.predict(X)
```

```python
data_xgb = data.copy()
```

```python
data_xgb['XGB_result'] = pred_xgb.copy()
```

```python
data_xgb['XGB_diff'] = (data_xgb['XGB_result'] - data_xgb['RENT_SQFT'])/data_xgb['RENT_SQFT']
```

```python
data_xgb.to_csv('xgb.csv')
```

```python
mean_value = train_X['PROPERTY_ACRES'].mean()
```

```python
train_X['PROPERTY_ACRES'].fillna(value = mean_value, inplace = True)
```

```python
test_X['PROPERTY_ACRES'].fillna(value = test_X['PROPERTY_ACRES'].mean(), inplace = True)
```

```python
from sklearn.linear_model import ElasticNet
gsc_enet = GridSearchCV(
         estimator = ElasticNet(),
         param_grid = {'alpha': [0.0001, 0.001, 0.01],
                       'l1_ratio': np.arange(0, 1, 0.02),
                       'max_iter': [80, 200, 500],
                       'tol': [0.01, 0.1, 0.3]
                    },
         cv = 3,
         scoring ='neg_root_mean_squared_error', n_jobs=3)

grid_result_enet = gsc_enet.fit(train_X, train_y)
best_params_enet = grid_result_enet.best_params_
```

```python
best_params_enet
```

```python
enet = ElasticNet(alpha=best_params_enet["alpha"],
                  l1_ratio=best_params_enet['l1_ratio'], 
                  max_iter=best_params_enet["max_iter"],
                  tol = best_params_enet["tol"])
enet.fit(train_X, train_y.values.ravel())
```

```python
enet_pre_train = enet.predict(train_X)
rmse_enet = np.sqrt(mean_squared_error(train_y, enet_pre_train))
```

```python
rmse_enet
```

```python
enet_pre_test = enet.predict(test_X)
rmse_enet_test = np.sqrt(mean_squared_error(test_y, enet_pre_test))
rmse_enet_test
```

```python
data_enet = data.copy()
mean_value = data_enet['PROPERTY_ACRES'].mean()
data_enet['PROPERTY_ACRES'].fillna(value = mean_value, inplace = True)
pred_enet = enet.predict(X)
```

```python
data_enet['ENET_result'] = pred_enet.copy()
```

```python
data_enet['ENET_diff'] = (data_enet['ENET_result'] - data_enet['RENT_SQFT'])/data_enet['RENT_SQFT']
```

```python
data_enet.to_csv('enet.csv')
```
